## 1) Descripción del problema
En un centro de atención médica se requiere un sistema para gestionar pacientes usando **estructuras de datos lineales implementadas desde cero** (sin `collections`, sin `deque`, sin librerías externas).

- **Pacientes normales (N):** se atienden por orden de llegada (FIFO).
- **Pacientes críticos (C):** tienen prioridad sobre los normales, pero entre críticos también se respeta FIFO.
- **Atención:** solo puede atenderse un paciente a la vez y, al atenderlo, se elimina del sistema.

Además, se solicita implementar dos estructuras base:

1. **Pila (LIFO):** métodos `push(elemento)`, `pop()`, `top()`, `esta_vacia()`, `tamaño()`.
2. **Cola (FIFO):** métodos `encolar(elemento)`, `desencolar()`, `frente()`, `esta_vacia()`, `tamaño()`.

El objetivo del sistema es procesar `n` acciones de entrada y producir la salida exacta según el enunciado del laboratorio.

## 2) Requerimientos
### Requerimientos funcionales

- **RF2.** El sistema debe permitir implementar una estructura tipo Cola con comportamiento FIFO (primero en entrar, primero en salir).
- **RF3.** El sistema debe gestionar dos colas de pacientes (normales y críticos) y un historial de atendidos.
- **RF4.** El sistema debe permitir registrar pacientes normales con la acción `N nombre` y mostrar `Paciente N nuevo : nombre`.

## 7) Análisis de complejidad

### 7.1) Resumen del problema
El sistema procesa `n` acciones de entrada (`N nombre`, `C nombre`, `A`) usando: Una cola de pacientes críticos, una cola de pacientes normales y una pila para historial de atendidos. En cada iteración se ejecuta una operación principal sobre estructuras lineales, sin ciclos anidados. Por eso, el comportamiento global es lineal respecto a `n`.

### 7.2) Tabla compacta de complejidad

| Componente | Operación principal | Tiempo | Espacio adicional |
|---|---|---|---|
| `Pila` | `push`, `pop`, `top`, `esta_vacia`, `tamaño` | $O(1)$ (amortizado en `push`) | $O(k)$ |
| `Cola` | `encolar`, `desencolar`, `frente`, `esta_vacia`, `tamaño` | $O(1)$ amortizado | $O(k)$ |
| `SistemaAtencionMedica` | `agregar_paciente`, `atender_paciente`, `procesar_accion` | $O(1)$ por acción (excepto parseo de línea) | $O(1)$ |

Donde `k` es la cantidad de elementos almacenados en la estructura.

### 7.3) Complejidad total

Sea `n` el número de acciones:

$$
T(n)=\sum_{i=1}^{n} O(1)=O(n)
$$

| Escenario | Complejidad temporal | Complejidad espacial |
|---|---|---|
| Mejor caso (acciones de costo constante) | $\Theta(n)$ | $\Theta(1)$ |
| Peor caso (muchas inserciones sin atender) | $\Theta(n)$ | $\Theta(n)$ |
| Caso promedio | $\Theta(n)$ | $O(n)$ |

## 9) Código funcional
Implementación completa del laboratorio con:

- Estructura `Pila` (LIFO).
- Estructura `Cola` (FIFO, sin `collections` ni `deque`).
- Sistema de atención médica con prioridad de críticos.
- Lectura de `n` acciones por entrada estándar.
- Salida exacta según el formato solicitado.

```python
# Parte 1: Implementación de estructuras
class Pila:
    def __init__(self):
        self._datos = []

    def push(self, elemento):
        self._datos.append(elemento)

    def pop(self):
        if self.esta_vacia():
            return None
        return self._datos.pop()

    def top(self):
        if self.esta_vacia():
            return None
        return self._datos[-1]

    def esta_vacia(self):
        return len(self._datos) == 0

    def tamaño(self):
        return len(self._datos)


class Cola:
    def __init__(self):
        self._datos = []

    def encolar(self, elemento):
        self._datos.append(elemento)

    def desencolar(self):
        if self.esta_vacia():
            return None
        # pop(0) es la forma directa de implementar FIFO sin deque
        return self._datos.pop(0)

    def frente(self):
        if self.esta_vacia():
            return None
        return self._datos[0]

    def esta_vacia(self):
        return len(self._datos) == 0

    def tamaño(self):
        return len(self._datos)


# Parte 2: Sistema de Atención Médica
class SistemaAtencionMedica:
    def __init__(self):
        self.cola_criticos = Cola()
        self.cola_normales = Cola()

    def agregar_paciente(self, tipo, nombre):
        if tipo == "C":
            self.cola_criticos.encolar(nombre)
            return f"Paciente C nuevo : {nombre}"
        elif tipo == "N":
            self.cola_normales.encolar(nombre)
            return f"Paciente N nuevo : {nombre}"
        return "accion invalida"

    def atender_paciente(self):
        # Regla: Críticos primero, luego normales
        if not self.cola_criticos.esta_vacia():
            paciente = self.cola_criticos.desencolar()
            return f"Se atiende a {paciente}"
        
        if not self.cola_normales.esta_vacia():
            paciente = self.cola_normales.desencolar()
            return f"Se atiende a {paciente}"
        
        return "no hay pacientes"

def main():
    sistema = SistemaAtencionMedica()
    
    linea_n = input().strip()
    if not linea_n or not linea_n.isdigit():
        return
    
    n = int(linea_n)
    for _ in range(n):
        linea = input().strip()
        if not linea:
            continue
            
        partes = linea.split(maxsplit=1)
        comando = partes[0]

        if comando == "A":
            print(sistema.atender_paciente())
        elif comando in ("N", "C") and len(partes) == 2:
            print(sistema.agregar_paciente(comando, partes[1]))
        else:
            print("no hay pacientes" if comando == "A" else "accion invalida")

# Ejecución directa solicitada
main()
```
