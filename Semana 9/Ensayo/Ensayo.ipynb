## Resumen

El presente ensayo constituye un análisis académico extenso y profundo sobre cinco de las estructuras de datos más relevantes en el diseño de sistemas computacionales contemporáneos: la **Pila** (*Stack*), la **Cola** (*Queue*), el **Árbol Binario de Búsqueda** (BST, del inglés *Binary Search Tree*), el **Árbol AVL** —denominado así por sus creadores Adelson-Velsky y Landis— y el **Grafo**, estructura no lineal que modela relaciones complejas entre entidades. Estas estructuras fueron estudiadas en las semanas 7, 8 y 9 del curso de Diseño de Datos y Algoritmos, y su análisis integrado revela una progresión natural desde la simplicidad lineal hacia la sofisticación jerárquica y reticular.

Durante la **semana 7**, el estudio se centró en las estructuras de datos lineales fundamentales. La **Pila**, cuyo principio de operación es LIFO (*Last In, First Out*), constituye la base de múltiples mecanismos en informática: desde la gestión del *call stack* durante la ejecución de programas, hasta la implementación de operaciones de deshacer y rehacer en editores de texto o la evaluación de expresiones matemáticas en compiladores. La **Cola**, regida por el principio FIFO (*First In, First Out*), sustenta los sistemas de gestión de turnos, impresoras en red, buffers de comunicación y, como se evidenció en el laboratorio de la semana, sistemas de atención médica con prioridades. La implementación de ambas estructuras desde cero —sin librerías externas— permitió comprender sus mecanismos internos, sus fortalezas y, crucialmente, sus limitaciones: el `desencolar()` implementado con `list.pop(0)` incurre en un costo de $O(n)$ por desplazamiento de elementos, transformando una secuencia de $n$ operaciones en un proceso de complejidad cuadrática $O(n^2)$. Esta limitación motiva naturalmente el estudio de estructuras con mayor sofisticación.

La **semana 8** introdujo el concepto de árbol como estructura jerárquica no lineal, centrándose en el **Árbol Binario de Búsqueda** (BST) y el **Árbol AVL**. El BST organiza los datos respetando la propiedad de ordenamiento: los valores del subárbol izquierdo son menores que la raíz, y los del derecho son mayores. Esto permite búsquedas, inserciones y eliminaciones en tiempo promedio de $O(\log n)$. Sin embargo, el BST presenta una vulnerabilidad crítica ante la llegada de datos ordenados, que lo degenera en una lista enlazada de altura $O(n)$, anulando todas sus ventajas algorítmicas. El **Árbol AVL** resuelve esta fragilidad estructural mediante un mecanismo de **autobalanceo automático**: tras cada inserción, el árbol detecta desbalanceos mediante factores de altura y aplica rotaciones (*LL*, *RR*, *LR* o *RL*) para restaurar el equilibrio. El resultado es una garantía matemática de altura $O(\log n)$ en todo escenario, haciendo de las operaciones un proceso consistente y predecible. El laboratorio de inventario de computadores demostró de manera práctica cómo la elección entre BST y AVL puede significar la diferencia entre un sistema eficiente y uno que colapsa ante patrones de datos ordenados.

La **semana 9** amplió el horizonte hacia las estructuras de datos no lineales de mayor complejidad: los **grafos**. Un grafo es un par ordenado $G = (V, E)$ donde $V$ es un conjunto de vértices y $E$ es un conjunto de aristas que representan relaciones entre ellos. Los grafos son el modelo matemático más general y poderoso disponible en ciencias de la computación: redes de distribución, redes sociales, mapas de rutas, internet y circuitos eléctricos son todos instancias de grafos. Se estudiaron dos algoritmos de recorrido esenciales: **BFS** (*Breadth-First Search*), que explora nivel por nivel garantizando el camino con menor número de saltos en grafos no ponderados, y **DFS** (*Depth-First Search*), que profundiza por cada rama antes de retroceder, siendo ideal para la detección de ciclos y la exploración exhaustiva. Ambos algoritmos presentan complejidad temporal $O(V + E)$, siendo eficientes incluso en grafos de gran escala.

En conjunto, el estudio de estas cinco estructuras describe una narrativa coherente sobre el diseño de algoritmos: cada estructura ofrece una solución óptima bajo condiciones específicas, y su correcta selección es el primer y más determinante paso para construir sistemas computacionales eficientes, escalables y robustos. Este ensayo analiza con profundidad cada estructura a través de sus fundamentos teóricos, sus aplicaciones prácticas, plantillas de diseño de código, sus ventajas y desventajas, y establece una comparación global que permite al lector comprender cuándo y por qué elegir cada una de ellas.

## 2. Semana 7: Estructuras de Datos Lineales — Pilas y Colas

### 2.1 Fundamentos Teóricos

Las **estructuras de datos lineales** son aquellas en las cuales los elementos se organizan en una secuencia única, donde cada elemento tiene como máximo un predecesor y un sucesor directo. Son la forma más intuitiva de organizar datos y la puerta de entrada al estudio formal de las estructuras de datos. Entre ellas, la **Pila** (*Stack*) y la **Cola** (*Queue*) son las más fundamentales y, por extensión, las más universalmente aplicadas en la ingeniería del software.

La diferencia esencial entre una Pila y una Cola no reside en la estructura interna de almacenamiento —ambas pueden implementarse sobre un arreglo dinámico o una lista enlazada— sino en la **disciplina de acceso**: la política que determina qué elemento puede ser extraído en cada momento. Esta política es, en esencia, lo que confiere a cada estructura su identidad algorítmica, su comportamiento predecible y sus casos de uso específicos.

---

### 2.2 La Pila (*Stack*) — Principio LIFO

Una **Pila** es una estructura de datos que opera bajo el principio **LIFO** (*Last In, First Out*: el último elemento en ingresar es el primero en salir). La analogía física es irresistible: imagine una pila de platos en una cocina; solo se puede agregar o retirar del tope, y los elementos del fondo permanecen inaccesibles hasta vaciar todos los superiores.

#### Propiedades fundamentales de la Pila

| Propiedad | Descripción |
|:---|:---|
| **Acceso restringado al tope** | Solo el elemento en la posición superior (*top*) es directamente accesible |
| **Inserción al tope** | `push(elemento)` coloca el nuevo elemento siempre en la cima |
| **Extracción del tope** | `pop()` elimina y retorna el elemento de la cima; la Pila se reduce en uno |
| **Consulta no destructiva** | `top()` permite leer el tope sin eliminarlo |
| **Verificación de vacío** | `esta_vacia()` permite consultar el estado sin riesgo de errores |

#### Complejidad temporal de operaciones

| Operación | Complejidad | Justificación |
|:---|:---:|:---|
| `push(elemento)` | $O(1)$ amortizado | Agrega al final del arreglo interno |
| `pop()` | $O(1)$ | Elimina el último elemento del arreglo |
| `top()` | $O(1)$ | Accede por índice al último elemento |
| `esta_vacia()` | $O(1)$ | Compara longitud con cero |
| `tamaño()` | $O(1)$ | Retorna longitud pre-calculada |

#### Aplicaciones canónicas de la Pila

Las pilas no son una abstracción académica; son el mecanismo subyacente en herramientas que todos los programadores usan a diario:

- **Gestión del *call stack*:** Cada vez que se llama a una función, el entorno de ejecución empuja un *marco de activación* (*stack frame*) con los parámetros y variables locales. Al retornar, ese marco es extraído. La Pila de llamadas es literalmente el modelo de ejecución de todos los lenguajes compilados e interpretados.
- **Algoritmo DFS iterativo:** El recorrido en profundidad de grafos puede implementarse con una pila explícita en lugar de recursión, evitando el riesgo de desbordamiento del *call stack* en grafos muy grandes.
- **Evaluación de expresiones matemáticas:** Los compiladores utilizan pilas para convertir expresiones en notación infija a notación postfija y evaluarlas sin ambigüedad de precedencia.
- **Historial de navegación:** Los botones "Atrás" y "Adelante" de un navegador web utilizan dos pilas para mantener el historial de páginas visitadas.
- **Verificación de paréntesis balanceados:** Un problema clásico de compiladores que se resuelve en una sola pasada $O(n)$ usando una pila.
- **Operación *Undo/Redo* en editores:** Cada acción del usuario se empuja a la pila de deshacer; al presionar Ctrl+Z, la acción del tope es revertida.

### 2.3 La Cola (*Queue*) — Principio FIFO

Una **Cola** opera bajo el principio **FIFO** (*First In, First Out*: el primer elemento en ingresar es el primero en salir). La analogía física es una fila de espera en una entidad bancaria: el primer cliente en llegar es el primero en ser atendido; nuevos clientes siempre se agregan al final de la fila. Este comportamiento modela de manera precisa innumerables escenarios de la vida real.

#### Propiedades fundamentales de la Cola

| Propiedad | Descripción |
|:---|:---|
| **Inserción al final** | `encolar(elemento)` agrega siempre al extremo trasero (*rear*) |
| **Extracción del frente** | `desencolar()` elimina y retorna siempre el elemento del extremo frontal (*front*) |
| **Consulta no destructiva** | `frente()` permite leer el próximo elemento sin extraerlo |
| **Equidad de servicio** | Todos los elementos son atendidos en el orden estricto de su llegada |

#### La trampa de `list.pop(0)` — Una lección de complejidad

La implementación ingenua de `desencolar()` usando `list.pop(0)` en Python incurre en un costo $O(n)$ por operación, pues el intérprete debe **desplazar todos los elementos restantes** una posición hacia el inicio del arreglo interno. Si se realizan $n$ operaciones de `desencolar` sobre una cola que crece hasta tener $n$ elementos, el costo acumulado es:

$$T(n) = \sum_{i=1}^{n} O(i) = O\!\left(\frac{n(n+1)}{2}\right) = O(n^2)$$

Esta degradación cuadrática, aparentemente inocua en implementaciones pequeñas de laboratorio, se vuelve catastrófica en sistemas de producción que procesan miles o millones de eventos. La solución óptima es usar `collections.deque`, que mantiene $O(1)$ en **ambos extremos** mediante una lista doblemente enlazada.

| Operación | `list.pop(0)` | `collections.deque` |
|:---|:---:|:---:|
| `encolar` | $O(1)$ amortizado | $O(1)$ |
| `desencolar` | $O(n)$ ← **degradación** | $O(1)$ |
| Costo total $n$ ops | $O(n^2)$ | $O(n)$ |

#### Aplicaciones canónicas de la Cola

- **Gestión de procesos en sistemas operativos:** El planificador de CPU organiza procesos listos en colas. Los sistemas operativos modernos utilizan múltiples colas de prioridades para diferentes clases de procesos.
- **Algoritmo BFS en grafos:** Fundamental para explorar grafos nivel por nivel (abordado en Semana 9).
- **Buffers de comunicación en redes:** Los paquetes de datos esperan su turno de transmisión en colas FIFO dentro de los routers.
- **Sistemas de atención con prioridades:** Como el sistema médico del laboratorio, que combina dos colas (críticos y normales) con una política de prioridad.
- **Impresión en red:** Los trabajos de impresión se encolan y procesan en el orden de llegada.

### 2.4 Cola con Prioridad: La Extensión del Concepto FIFO

El laboratorio de la semana 7 introdujo una extensión poderosa y directamente aplicable del concepto de Cola: la **cola con prioridad doble** para un sistema de atención médica. En este sistema:

- **Pacientes críticos (C):** tienen prioridad absoluta sobre cualquier paciente normal, independientemente de su hora de llegada.
- **Pacientes normales (N):** se atienden por orden estricto de llegada (FIFO estándar).
- **Dentro de su categoría:** el orden FIFO se preserva rigurosamente —un segundo paciente crítico espera detrás del primero.

La solución elegante fue mantener **dos colas independientes** (una para críticos y otra para normales) con una política de selección que siempre privilegia la cola de críticos cuando no está vacía. Esta arquitectura implementa de manera fiel el sistema de **Triage** médico: un protocolo real de gestión de emergencias hospitalarias, computarizado con apenas dos estructuras de datos elementales y una lógica condicional simple.

> Esta es quizás la lección más poderosa de la semana 7: las estructuras de datos no existen en el vacío. Su valor real emerge cuando se combinan creativamente para modelar comportamientos del mundo real con elegancia, eficiencia y claridad conceptual.

#### Impacto del diseño de desencolar: $O(n)$ vs $O(1)$

Para ilustrar la gravedad de la diferencia, considérese un sistema que procesa $n = 10\,000$ pacientes:

- **Con `list.pop(0)`:** $\approx \frac{10000^2}{2} = 50\,000\,000$ operaciones elementales.
- **Con `collections.deque`:** $\approx 10\,000$ operaciones elementales.

La diferencia es de **5000×** en favor de `deque`. Este no es un detalle de implementación menor; es la diferencia entre un sistema que responde en milisegundos y uno que requiere segundos. En sistemas hospitalarios reales que procesan miles de eventos diarios, esta degradación es inaceptable.

### 2.7 Ventajas y Desventajas de las Estructuras Lineales

#### Ventajas

| # | Ventaja | Descripción |
|:---:|:---|:---|
| 1 | **Simplicidad conceptual** | Fáciles de entender, implementar y depurar. No requieren conocimiento de punteros ni balanceo. |
| 2 | **Eficiencia en memória** | Almacenamiento contiguo en memoria (arreglo dinámico); favorable para la caché del procesador. |
| 3 | **Operaciones rápidas** | `push`, `pop`, `encolar` en $O(1)$; ideales para operaciones de alta frecuencia. |
| 4 | **Modelo mental intuitivo** | Mapean directamente a conceptos del mundo real (fila de espera, pila de platos). |
| 5 | **Composición poderosa** | Dos colas simples más una política de prioridad implementan un sistema de Triage completo. |

#### Desventajas

| # | Desventaja | Descripción |
|:---:|:---|:---|
| 1 | **Sin acceso aleatorio** | No es posible acceder al $k$-ésimo elemento sin extraer los anteriores. |
| 2 | **Degradación con `list.pop(0)`** | La implementación ingenua de `desencolar` produce $O(n^2)$ total, un peligro oculto. |
| 3 | **Sin ordenamiento** | No organizan datos por valor; solo por orden de inserción. |
| 4 | **Sin búsqueda eficiente** | Encontrar un elemento específico requiere $O(n)$ en el caso general. |
| 5 | **Sin prioridad nativa** | La Cola básica no soporta prioridades; requiere extensiones como la doble-cola o Min-Heap. |

> **Cuándo usar una Pila:** cuando el problema exhibe estructura LIFO natural —llamadas recursivas, historial de acciones, retroceso en laberintos, evaluación de expresiones.
>
> **Cuándo usar una Cola:** cuando el problema requiere equidad de servicio FIFO —despacho de eventos, simulaciones, algoritmos de grafos como BFS, buffers de comunicación.

## 3. Semana 8: Estructuras Jerárquicas — Árbol BST y Árbol AVL

### 3.1 De lo Lineal a lo Jerárquico: El Salto Algorítmico

Las estructuras lineales estudiadas en la semana 7 ofrecen eficiencia en inserción y extracción ordenada, pero presentan una limitación fundamental: la **búsqueda de un elemento específico** requiere $O(n)$ en el caso general, pues no existe una organización que permita descartar subconjuntos del espacio de búsqueda. Superar esta barrera requiere adoptar una organización **jerárquica** de los datos.

Los **árboles** son estructuras de datos no lineales compuestas por **nodos** conectados por **aristas** de manera que no existen ciclos. Cada árbol tiene exactamente un nodo especial llamado **raíz**, desde el cual se puede alcanzar cualquier otro nodo siguiendo las conexiones. Un árbol binario es aquel en el que cada nodo tiene como máximo dos hijos: el hijo **izquierdo** y el hijo **derecho**.

---

### 3.2 El Árbol Binario de Búsqueda (BST)

Un **Árbol Binario de Búsqueda** es un árbol binario que satisface en todo momento la siguiente invariante estructural, conocida como la **propiedad BST**:

> Para todo nodo $x$ en el árbol:
> - Todos los valores en el **subárbol izquierdo** de $x$ son estrictamente **menores** que el valor de $x$.
> - Todos los valores en el **subárbol derecho** de $x$ son estrictamente **mayores** que el valor de $x$.

Esta propiedad, aplicada recursivamente, transforma el proceso de búsqueda en un problema de decisión binaria en cada nodo: ¿el valor buscado es menor o mayor que el nodo actual? En cada paso se descarta la mitad del árbol restante, produciendo búsquedas en $O(\log n)$ cuando el árbol está equilibrado.

#### Estructura de un nodo BST

Un nodo en el BST almacena:
- El **valor** (clave de búsqueda)
- Referencia al **hijo izquierdo** (o `None`)
- Referencia al **hijo derecho** (o `None`)

#### Operaciones fundamentales del BST

| Operación | Descripción | Complejidad promedio | Complejidad peor caso |
|:---|:---|:---:|:---:|
| `insertar(valor)` | Ubica la posición correcta según la propiedad BST e inserta | $O(\log n)$ | $O(n)$ |
| `buscar(valor)` | Navega el árbol comparando en cada nodo; retorna True/False | $O(\log n)$ | $O(n)$ |
| `inorden()` | Recorre izquierda → raíz → derecha; produce valores **ordenados** | $O(n)$ | $O(n)$ |
| `preorden()` | Recorre raíz → izquierda → derecha; útil para copiar el árbol | $O(n)$ | $O(n)$ |
| `postorden()` | Recorre izquierda → derecha → raíz; útil para eliminar el árbol | $O(n)$ | $O(n)$ |
| `altura()` | Cuenta la profundidad máxima desde la raíz hasta una hoja | $O(n)$ | $O(n)$ |

> **La propiedad más poderosa del BST:** el recorrido **Inorden** produce siempre los elementos en **orden ascendente**, sin necesidad de un algoritmo de ordenamiento separado. Este es el fundamento de los índices de bases de datos.

#### El talón de Aquiles del BST: el árbol degenerado

La fortaleza del BST tiene una condición crítica: las garantías de $O(\log n)$ solo se sostienen cuando el árbol está suficientemente **equilibrado**. Si los datos llegan en orden estrictamente creciente o decreciente —un escenario perfectamente posible en aplicaciones reales— el BST degenera en una **lista enlazada**, con todos los nodos en una única cadena lateral:

```
[10]
  └─ D: [20]
         └─ D: [30]
                └─ D: [40]
                        └─ D: [50]
```

En este árbol degenerado, la altura es $h = n$, y **todas las operaciones se degradan a $O(n)$**: buscar el elemento mayor requiere atravesar todos los nodos. El laboratorio demostró este escenario al insertar `10, 20, 30, 40, 50` en un BST vacío, produciendo exactamente esta cadena lineal.

---

### 3.3 El Árbol AVL: la Solución al Desbalanceo

El **Árbol AVL**, propuesto por Adelson-Velsky y Landis en 1962, resuelve la debilidad del BST mediante un mecanismo de **autobalanceo automático**. La idea central es brillante en su simplicidad: mantener siempre una restricción sobre la diferencia de alturas entre subárboles.

#### Definición del Factor de Balance

Para cada nodo $x$ del árbol AVL, se define el **factor de balance** como:

$$FB(x) = h_{\text{izquierda}}(x) - h_{\text{derecha}}(x)$$

La invariante del AVL exige que en **todo** nodo del árbol:

$$|FB(x)| \leq 1 \quad \Rightarrow \quad FB(x) \in \{-1,\; 0,\; 1\}$$

Cuando una inserción viola esta invariante en algún nodo, el árbol aplica automáticamente una **rotación** para restaurar el equilibrio.

#### Las Cuatro Rotaciones del AVL

| Tipo | Caso | Desbalance | Solución |
|:---:|:---|:---:|:---|
| **LL** | Inserción en **subárbol izquierdo del hijo izquierdo** | $FB = +2$ | Rotación simple a la **derecha** sobre la raíz del desbalance |
| **RR** | Inserción en **subárbol derecho del hijo derecho** | $FB = -2$ | Rotación simple a la **izquierda** sobre la raíz del desbalance |
| **LR** | Inserción en **subárbol derecho del hijo izquierdo** | $FB = +2$ | Rotación doble: primero **izquierda** en el hijo, luego **derecha** en la raíz |
| **RL** | Inserción en **subárbol izquierdo del hijo derecho** | $FB = -2$ | Rotación doble: primero **derecha** en el hijo, luego **izquierda** en la raíz |

#### Ejemplo concreto de rotación RR (caso del laboratorio)

Al insertar `10 → 20 → 30` en un AVL vacío, ocurre el siguiente desbalance y corrección:

```
Paso 1: insertar 10
    [10]          FB(10) = 0  ✓

Paso 2: insertar 20
    [10]          FB(10) = -1  ✓
      └─D: [20]   FB(20) = 0

Paso 3: insertar 30 → DESBALANCE RR en nodo [10]
    [10]  FB = -2  ✗
      └─D: [20]
              └─D: [30]

Rotación izquierda (RR) en nodo [10]:
    [20]  ← nueva raíz
   /    \
[10]    [30]    todos los FB = 0  ✓
```

El resultado es un árbol perfectamente balanceado, y la altura se reduce de 3 a 2.

#### Garantía matemática del AVL

La altura de un Árbol AVL con $n$ nodos satisface:

$$h \leq 1.44 \cdot \log_2(n + 2) - 0.328 \approx O(\log n)$$

Esta cota garantiza que **toda operación de búsqueda, inserción y eliminación es $O(\log n)$**, independientemente del orden en que lleguen los datos. No existe un orden de inserción que pueda degradar el AVL.

#### Template 1: Función `bst_insertar` (BST recursivo)

```python
def bst_insertar(raiz, valor):
    """
    Inserta 'valor' en el BST cuya raíz es 'raiz'.
    Retorna la nueva raíz (puede ser el mismo nodo o uno nuevo).

    Propiedad BST:
      - valor < raiz['valor']  → insertar en subárbol izquierdo
      - valor > raiz['valor']  → insertar en subárbol derecho
      - valor == raiz['valor'] → ignorar (sin duplicados)

    Complejidad:
      - Tiempo: O(h) donde h = altura del árbol
        · Árbol balanceado: O(log n)
        · Árbol degenerado: O(n)
      - Espacio: O(h) en la pila de recursión
    """
    if raiz is None:
        return {'valor': valor, 'izq': None, 'der': None}

    if valor < raiz['valor']:
        raiz['izq'] = bst_insertar(raiz['izq'], valor)   # recorrer izquierda
    elif valor > raiz['valor']:
        raiz['der'] = bst_insertar(raiz['der'], valor)   # recorrer derecha
    # Si valor == raiz['valor']: no hacer nada (sin duplicados)
    return raiz
```

#### Template 2: Función `bst_buscar` (iterativo)

```python
def bst_buscar(raiz, valor):
    """
    Busca 'valor' en el BST. Retorna True si existe, False si no.

    Estrategia: en cada nodo, la propiedad BST permite decidir
    en qué subárbol continuar, descartando el otro completamente.

    Complejidad:
      - Tiempo: O(h) — O(log n) balanceado, O(n) degenerado
      - Espacio: O(1) — iterativo, sin pila de recursión
    """
    nodo_actual = raiz
    while nodo_actual is not None:
        if valor == nodo_actual['valor']:
            return True
        elif valor < nodo_actual['valor']:
            nodo_actual = nodo_actual['izq']   # ir a la izquierda
        else:
            nodo_actual = nodo_actual['der']   # ir a la derecha
    return False
```

#### Condición de degradación del BST

El BST alcanza su peor caso cuando los datos se insertan en **orden estrictamente creciente o decreciente**. Esto es significativamente frecuente en aplicaciones reales: identificadores auto-incrementales (IDs de base de datos), timestamps cronológicos, códigos de inventario generados secuencialmente. En todos estos casos, el BST produce un árbol degenerado que anula completamente su ventaja sobre una lista enlazada.

El AVL nunca degrada porque su mecanismo de rotación mantiene $h \leq 1.44 \cdot \log_2 n$, sin importar el orden de llegada de los datos.

### 3.7 Ventajas y Desventajas del BST y AVL

#### Ventajas del BST

| # | Ventaja | Descripción |
|:---:|:---|:---|
| 1 | **Búsqueda eficiente** | $O(\log n)$ en árbol balanceado; radicalmente mejor que $O(n)$ lineal |
| 2 | **Inorden produce orden** | El recorrido inorden entrega datos ordenados sin algoritmo adicional |
| 3 | **Inserción flexible** | Nuevos elementos se ubican en la posición correcta automáticamente |
| 4 | **Sin desperdicio de espacio** | Cada nodo ocupa exactamente el espacio necesario; no requiere arreglo contiguo |
| 5 | **Estructura dinámica** | Crece y decrece libremente, sin necesidad de redimensionar |

#### Desventajas del BST

| # | Desventaja | Descripción |
|:---:|:---|:---|
| 1 | **Degradación ante datos ordenados** | Insertar datos ordenados produce árbol lineal con $O(n)$ por operación |
| 2 | **Sin garantía de balance** | El rendimiento es impredecible; depende del orden de los datos |
| 3 | **Implementación más compleja que lineal** | Requiere manejo de punteros/referencias y navegación recursiva |
| 4 | **Eliminación compleja** | Eliminar un nodo con dos hijos requiere encontrar el sucesor inorden |

#### Ventajas del AVL

| # | Ventaja | Descripción |
|:---:|:---|:---|
| 1 | **Rendimiento garantizado** | $O(\log n)$ para inserción, búsqueda y eliminación en **todo** escenario |
| 2 | **Robusto ante datos ordenados** | Inserciones en cualquier orden producen un árbol balanceado automáticamente |
| 3 | **Altura logarítmica estricta** | $h \leq 1.44 \cdot \log_2(n+2)$ — eficiencia máxima de búsqueda binaria |
| 4 | **Fundamento de bases de datos** | Los índices B-Tree de bases de datos como PostgreSQL y MySQL se inspiran en árboles autobalanceados |

#### Desventajas del AVL

| # | Desventaja | Descripción |
|:---:|:---|:---|
| 1 | **Mayor complejidad de implementación** | Las rotaciones agregan lógica adicional; depurar un AVL es más difícil |
| 2 | **Overhead por rotaciones** | Aunque cada rotación es $O(1)$, añaden una constante de tiempo en inserciones |
| 3 | **Almacenamiento del factor de balance** | Cada nodo debe almacenar su altura o factor de balance como dato adicional |
| 4 | **Necesidad de un BST como base** | El AVL es un BST extendido; requiere dominar el BST previamente |

> **Cuándo usar BST:** datos con inserción aleatoria donde el balance natural es probable, o cuando la implementación debe ser mínimamente simple y los datos son conocidamente no ordenados.
>
> **Cuándo usar AVL:** datos con cualquier patrón de llegada, sistemas de producción donde el rendimiento predecible es crítico, o cualquier escenario donde los datos puedan llegar en orden (IDs, timestamps, códigos secuenciales).

## 4. Semana 9: Estructuras de Red — Grafos y Algoritmos de Recorrido

### 4.1 De lo Jerárquico a lo General: La Explosión del Modelado

Los árboles BST y AVL imponen una restricción estructural severa: **todo nodo tiene exactamente un padre y a lo sumo dos hijos**. Esta jerarquía estricta los hace poderosos para búsqueda ordenada, pero los vuelve incapaces de modelar relaciones donde una entidad puede estar conectada con múltiples otras de forma arbitraria —sin jerarquía, sin raíz, con ciclos posibles.

Los **grafos** eliminan todas estas restricciones. Un grafo es la estructura de datos más general disponible en ciencias de la computación: puede modelar cualquier sistema de relaciones entre entidades, independientemente de su complejidad topológica. Internet, el sistema nervioso, las redes sociales, las rutas aéreas, los circuitos eléctricos, los mapas de calles y las dependencias entre módulos de software son todos grafos.

---

### 4.2 Fundamentos de la Teoría de Grafos

#### Definición formal

Un **grafo** $G$ se define formalmente como un par ordenado:

$$G = (V, E)$$

donde:
- $V$ es un conjunto finito de **vértices** (también llamados nodos) — representan las entidades.
- $E \subseteq V \times V$ es un conjunto de **aristas** (también llamadas arcos o bordes) — representan las relaciones entre entidades.

#### Clasificación de grafos

| Clasificación | Criterio | Descripción |
|:---|:---|:---|
| **No dirigido** | Aristas sin dirección | Si $(u, v) \in E$, entonces $(v, u) \in E$ también. Las relaciones son simétricas. |
| **Dirigido** (*Dígrafo*) | Aristas con dirección | $(u, v) \in E$ no implica $(v, u) \in E$. Las relaciones pueden ser asimétricas. |
| **Ponderado** | Aristas con peso | Cada arista lleva un valor numérico (distancia, costo, capacidad). |
| **No ponderado** | Aristas sin peso | Solo importa si existe o no la conexión. |
| **Conexo** | Propiedad de alcanzabilidad | Existe un camino entre cualquier par de vértices. |
| **Con ciclos** | Presencia de ciclos | Es posible partir de un nodo y regresar a él por un camino diferente. |

El laboratorio de la semana 9 trabajó con un **grafo no dirigido y no ponderado** —la red de bodegas de distribución— donde las aristas representan rutas directas entre pares de bodegas.

---

### 4.3 Representación: Lista de Adyacencia

Existen dos representaciones canónicas de un grafo: la **matriz de adyacencia** y la **lista de adyacencia**.

**Matriz de adyacencia:** Una matriz $V \times V$ donde la posición $[i][j]$ vale 1 si existe arista entre el vértice $i$ y el $j$, y 0 en caso contrario. Es eficiente para consultas de existencia de arista en $O(1)$, pero ocupa $O(V^2)$ de espacio, inaceptable para grafos dispersos.

**Lista de adyacencia:** Un diccionario (o arreglo) donde la clave es cada vértice y el valor es la lista de sus vecinos directos. Es eficiente en espacio: $O(V + E)$, ideal para grafos dispersos donde $E \ll V^2$.

$$\text{Espacio:} \quad \text{Matriz} = O(V^2) \quad \text{vs} \quad \text{Lista} = O(V + E)$$

Para el grafo de 8 bodegas con 9 aristas, la lista de adyacencia ocupa espacio proporcional a $8 + 9 \times 2 = 26$ entradas, mientras que la matriz requeriría $8^2 = 64$ celdas —más del doble de espacio, con la mayoría en cero.

---

### 4.4 BFS — Búsqueda en Anchura (*Breadth-First Search*)

El algoritmo BFS explora el grafo **nivel por nivel**: primero visita todos los vecinos directos del nodo inicial (nivel 1), luego todos los vecinos de los vecinos que aún no han sido visitados (nivel 2), y así sucesivamente. Utiliza una **cola** como estructura auxiliar para garantizar este orden de exploración.

#### Algoritmo BFS — Descripción detallada

1. Inicializar una **cola** con el nodo de inicio; marcarlo como visitado.
2. Mientras la **cola no esté vacía**:
   a. Extraer el primer nodo de la cola (operación `desencolar`).
   b. Procesarlo (agregarlo al orden de visita).
   c. Por cada vecino **no visitado** del nodo actual, en orden alfabético: marcarlo como visitado y encolarlo.
3. Cuando la cola se vacía, todos los nodos alcanzables han sido visitados.

#### Propiedades fundamentales de BFS

| Propiedad | Descripción |
|:---|:---|
| **Orden de visita** | Nivel por nivel — primero todos los vecinos directos, luego sus vecinos, etc. |
| **Camino más corto** | BFS garantiza encontrar el camino con **menor número de aristas** entre dos nodos |
| **Completitud** | Si existe camino entre el nodo inicial y un nodo destino, BFS lo encontrará |
| **Complejidad temporal** | $O(V + E)$ — visita cada vértice y arista exactamente una vez |
| **Complejidad espacial** | $O(V)$ — la cola puede contener en el peor caso todos los vértices |

#### Ejemplo en el grafo de bodegas (A–H)

Con aristas `A-B, A-C, B-D, B-E, C-F, C-G, D-H, E-F, G-H` y explorando vecinos en orden alfabético:

| Nivel | Nodos visitados | Cola al inicio del nivel |
|:---:|:---|:---|
| 0 | A | [A] |
| 1 | B, C | [B, C] |
| 2 | D, E, F, G | [D, E, F, G] |
| 3 | H | [H] |

**Recorrido BFS:** `[A, B, C, D, E, F, G, H]`

El nodo H se alcanza en el nivel 3, lo que significa que el camino más corto de A a H tiene 3 aristas.

---

### 4.5 DFS — Búsqueda en Profundidad (*Depth-First Search*)

El algoritmo DFS explora el grafo **profundizando por una rama antes de retroceder**: sigue un camino hasta no poder avanzar más (nodo sin vecinos no visitados), luego retrocede al nodo anterior y continúa por otra rama. Se implementa naturalmente de forma recursiva, utilizando el *call stack* como pila implícita.

#### Algoritmo DFS — Descripción detallada

1. Marcar el nodo actual como visitado; agregarlo al orden de visita.
2. Por cada vecino **no visitado** del nodo actual, en orden alfabético:
   a. Llamar recursivamente a DFS sobre ese vecino.
3. Cuando todos los vecinos del nodo están visitados, retroceder al nodo que llamó (retorno de la recursión).

#### Propiedades fundamentales de DFS

| Propiedad | Descripción |
|:---|:---|
| **Orden de visita** | Por ramas — profundiza completamente por la primera rama antes de explorar la segunda |
| **Detección de ciclos** | DFS puede detectar ciclos en $O(V + E)$ verificando si se llega a un nodo ya visitado |
| **Exploración exhaustiva** | Garantiza visitar todo el grafo conexo; útil para auditar conexiones |
| **Componentes conexas** | Ejecutar DFS desde cada nodo no visitado permite identificar todas las componentes |
| **Complejidad temporal** | $O(V + E)$ — misma asíntota que BFS |
| **Complejidad espacial** | $O(V)$ — la pila de recursión puede tener en el peor caso profundidad $V$ |

#### Ejemplo en el grafo de bodegas (A–H)

**Recorrido DFS** explorando vecinos en orden alfabético:

```
A → B (primera opción alfabética de A)
  B → D (primera opción no visitada de B)
    D → H (única opción no visitada de D)
      H → G (primera opción no visitada de H; D ya visitado)
        G → C (primera opción no visitada de G; H ya visitado)
          C → F (primera opción no visitada de C; A y G ya visitados)
            F → E (única opción no visitada de F; C ya visitado)
Retrocede hasta A; B ya visitado; C ya visitado → FIN
```

**Recorrido DFS:** `[A, B, D, H, G, C, F, E]`

A diferencia de BFS que produce una secuencia por niveles, DFS produce una secuencia que refleja la topología de las ramas del grafo.

---

### 4.6 BFS vs DFS: Cuándo Elegir Cada Uno

| Criterio | BFS | DFS |
|:---|:---:|:---:|
| **Ruta más corta** (menor número de saltos) |  Garantiza el óptimo |  No garantiza el mínimo |
| **Exploración exhaustiva** de todas las ramas |  No óptimo |  Natural y eficiente |
| **Detección de ciclos** |  Posible pero no natural |  Muy natural |
| **Componentes conexas** | ✅ | ✅ |
| **Grafos con ramas muy largas** | Mejor (no profundiza) |  Puede tener *stack overflow* |
| **Grafos muy anchos** (muchos vecinos por nodo) |  Cola muy grande | Mejor (no encola por nivel) |
| **Memoria utilizada** | $O(V)$ en cola (puede ser grande) | $O(h)$ en pila — $h$ la profundidad máxima |

> **Síntesis:** Si la pregunta es *"¿cuál es la ruta más corta?"*, usar BFS. Si la pregunta es *"¿existe algún camino / hay ciclos / cuántas componentes separadas?"*, usar DFS. Si se necesita explorar toda la red sin importar el orden, ambos son equivalentes en eficiencia.

### 4.7 Templates de Diseño: Grafo, BFS y DFS

---

#### Template 1: Creación y manipulación del Grafo (lista de adyacencia)

```python
def crear_grafo():
    """
    Crea un grafo vacío representado como diccionario (lista de adyacencia).
    Complejidad: O(1).
    """
    return {}

def agregar_vertice(grafo, vertice):
    """
    Agrega un vértice al grafo sin duplicarlo.
    Complejidad: O(1) amortizado — acceso y asignación en diccionario.
    Retorna True si fue agregado, False si ya existía.
    """
    if vertice not in grafo:
        grafo[vertice] = []
        return True
    return False

def agregar_arista(grafo, u, v):
    """
    Agrega arista no dirigida entre u y v (bidireccional).
    Verifica existencia de vértices y ausencia de duplicados.
    Complejidad: O(grado(u) + grado(v)) para verificar duplicados.
    Retorna True si fue agregada, False si ya existía o hay error.
    """
    if u == v or u not in grafo or v not in grafo:
        return False
    if v not in grafo[u]:
        grafo[u].append(v)
        grafo[v].append(u)
        return True
    return False

def imprimir_grafo(grafo):
    """
    Imprime la lista de adyacencia ordenada alfabéticamente.
    Complejidad: O(V + E) — itera sobre todos los vértices y vecinos.
    """
    for vertice in sorted(grafo):
        print(f"  {vertice} <-> {sorted(grafo[vertice])}")
```

---

#### Template 2: Algoritmo BFS

```python
from collections import deque

def bfs(grafo, inicio):
    """
    Recorre el grafo en anchura (nivel por nivel) desde 'inicio'.
    Utiliza una cola (deque) para garantizar el orden por niveles.

    Complejidad:
      - Tiempo:  O(V + E) — cada vértice y arista procesados exactamente una vez
      - Espacio: O(V)     — la cola y el conjunto de visitados
    """
    if inicio not in grafo:
        return []

    visitados = set()              # Conjunto de nodos ya procesados (O(1) lookup)
    cola = deque([inicio])         # Cola inicializada con el nodo de partida
    visitados.add(inicio)
    orden = []

    while cola:                    # Mientras haya nodos por procesar
        nodo = cola.popleft()      # Extraer el nodo al frente — O(1) con deque
        orden.append(nodo)

        for vecino in sorted(grafo[nodo]):    # Vecinos en orden alfabético
            if vecino not in visitados:
                visitados.add(vecino)
                cola.append(vecino)           # Encolar para procesamiento futuro

    return orden
```

---

#### Template 3: Algoritmo DFS (recursivo)

```python
def _dfs_recursivo(grafo, nodo, visitados, orden):
    """
    Función auxiliar recursiva del DFS.
    Profundiza por cada vecino no visitado, en orden alfabético.
    Complejidad: O(V + E) total al ser llamada desde dfs().
    """
    visitados.add(nodo)
    orden.append(nodo)

    for vecino in sorted(grafo[nodo]):     # Orden determinista
        if vecino not in visitados:
            _dfs_recursivo(grafo, vecino, visitados, orden)

def dfs(grafo, inicio):
    """
    Recorre el grafo en profundidad (por ramas) desde 'inicio'.
    Implementado de forma recursiva; el call stack actúa como pila implícita.

    Complejidad:
      - Tiempo:  O(V + E) — igual que BFS
      - Espacio: O(V)     — pila de recursión de profundidad máxima V
    Riesgo: RecursionError si el grafo es muy denso; considerar versión iterativa.
    """
    if inicio not in grafo:
        return []

    visitados = set()
    orden = []
    _dfs_recursivo(grafo, inicio, visitados, orden)
    return orden
```

---

#### Template 4: Inicialización del grafo de bodegas A–H

```python
def inicializar_grafo_bodegas():
    """
    Construye el grafo inicial con 8 bodegas (A–H) y
    las 9 aristas de la red de distribución.

    Aristas: A-B, A-C, B-D, B-E, C-F, C-G, D-H, E-F, G-H

    Nivel 0: A
    Nivel 1: B, C
    Nivel 2: D, E, F, G
    Nivel 3: H
    """
    grafo = crear_grafo()
    for v in ["A", "B", "C", "D", "E", "F", "G", "H"]:
        agregar_vertice(grafo, v)
    for u, v in [("A","B"),("A","C"),("B","D"),("B","E"),
                 ("C","F"),("C","G"),("D","H"),("E","F"),("G","H")]:
        agregar_arista(grafo, u, v)
    return grafo
```

> **Énfasis de diseño:** La separación entre la creación del grafo (funciones independientes para vértice y arista) y los algoritmos de recorrido (BFS, DFS) respeta el principio de cohesión funcional. Cada función hace exactamente una cosa; la composición de varias produce el sistema completo.

### 4.8 Aplicación Práctica: Red de Bodegas de Distribución

El laboratorio de la semana 9 aplicó los conceptos de grafos a un escenario empresarial concreto: una empresa de distribución con **8 bodegas** (A–H) conectadas por rutas directas. La red fue construida como un grafo no dirigido donde cada bodega es un vértice y cada ruta directa entre dos bodegas es una arista.

#### La red original (bodegas A–H)

```
          [A]
         /   \
       [B]   [C]
       / \   / \
     [D] [E][F] [G]
      |   |      |
     [H]  +------+
```

Con aristas: `A-B, A-C, B-D, B-E, C-F, C-G, D-H, E-F, G-H`

Los recorridos desde la bodega A arrojaron:
- **BFS:** `[A, B, C, D, E, F, G, H]` — explora por niveles de proximidad.
- **DFS:** `[A, B, D, H, G, C, F, E]` — profundiza por la rama izquierda completa antes de explorar la derecha.

#### Impacto de agregar la bodega I (conexiones I-E e I-H)

Al incorporar la bodega I con conexiones a E y H, la red aumenta a $V = 9$, $E = 11$:

| Recorrido | Antes (A–H) | Después (A–I) |
|:---:|:---|:---|
| BFS desde A | `[A, B, C, D, E, F, G, H]` | `[A, B, C, D, E, F, G, H, I]` |
| DFS desde A | `[A, B, D, H, G, C, F, E]` | `[A, B, D, H, G, C, F, E, I]` |

El nodo I es alcanzado como **último nodo** en ambos recorridos porque sus conexiones (E y H) son nodos de profundidad 2 y 3 respectivamente en el grafo original. Esto demuestra que la topología del grafo determina el orden en que se descubren los nodos, y que cambios estructurales pueden alterar significativamente los caminos de recorrido.

#### Valor empresarial del modelado con grafos

El sistema de rutas de distribución modelado como grafo permite:

1. **BFS desde cualquier bodega:** determina cuál es la bodega más próxima (en saltos) para atender un pedido urgente.
2. **DFS sobre la red completa:** verifica que todas las bodegas son alcanzables (grafo conexo), lo que garantiza que no existen bodegas aisladas.
3. **Detección de atajos:** al agregar la bodega I entre E y H, se crea un nuevo camino alternativo que puede reducir el número de saltos en ciertas rutas.

### 4.9 Análisis de Complejidad — Grafos y Recorridos

#### Tabla de complejidades (Semana 9)

| Operación | Función | Tiempo | Espacio | Observación |
|:---|:---|:---:|:---:|:---|
| Crear grafo | `crear_grafo` | $O(1)$ | $O(1)$ | Diccionario vacío |
| Agregar vértice | `agregar_vertice` | $O(1)$ amortizado | $O(1)$ | Acceso a diccionario |
| Agregar arista | `agregar_arista` | $O(\deg(u))$ | $O(1)$ | Verificación de duplicado |
| Imprimir grafo | `imprimir_grafo` | $O(V + E)$ | $O(V)$ | Itera vértices y vecinos |
| BFS completo | `bfs` | $O(V + E)$ | $O(V)$ | Cola + visitados |
| DFS completo | `dfs` | $O(V + E)$ | $O(V)$ | Pila de recursión |
| Ordenar vecinos por exploración | `sorted(grafo[nodo])` | $O(d \log d)$ por nodo | $O(d)$ | $d$ = grado del nodo |

Dado que $\sum_{v \in V} d(v) = 2E$ (cada arista contribuye a dos listas), el costo total de ordenar todos los vecinos es $O(E \log d_{\max})$, que para grafos dispersos con grado máximo pequeño y constante resulta en $O(E)$.

$$T_{BFS} = T_{DFS} = O(V + E)$$

### 5.2 Jerarquía de Potencia Expresiva

Las estructuras estudiadas forman una jerarquía natural de potencia expresiva:

$$\text{Pila} \subset \text{Cola} \subset \text{Árbol BST} \subset \text{Árbol AVL} \subset \text{Grafo}$$

- Toda Pila puede modelarse como un caso especial de Cola con política LIFO.
- Todo árbol es un grafo sin ciclos y con exactamente un nodo raíz.
- Todo BST es un árbol con la propiedad de ordenamiento.
- Todo AVL es un BST con la invariante de balance adicional.
- El Grafo es el caso más general: sin restricciones de jerarquía, ciclos ni número de conexiones.

Sin embargo, esta jerarquía de expresividad no implica que una estructura sea siempre superior a otra: **cada una es óptima en su dominio de aplicación**. Usar un Grafo para implementar una pila sería un desperdicio de complejidad; usar una Pila para modelar una red de distribución sería imposible.

### 6.1 La Importancia Fundamental de Estas Estructuras

El conjunto de estructuras de datos estudiado en las semanas 7, 8 y 9 no es una selección arbitraria de tópicos académicos: constituye el **núcleo central** del pensamiento algorítmico que subyace en todos los sistemas de software de escala real.

**Las Pilas y Colas** son omnipresentes en la infraestructura invisible que soporta la computación moderna. Cada vez que un navegador web retrocede a la página anterior, ejecuta una operación `pop()` sobre su pila de historial. Cada vez que múltiples usuarios envían solicitudes simultáneas a un servidor web, esas solicitudes son encoladas y procesadas en orden. Cada llamada a función en cualquier lenguaje de programación empuja un marco en el *call stack*. La aparente simplicidad de estas estructuras oculta una omnipresencia profunda.

**Los Árboles BST y AVL** son el fundamento de los sistemas de bases de datos, que son el corazón de cualquier aplicación empresarial moderna. Los índices de MySQL, PostgreSQL, Oracle y SQLite se implementan mediante B-Trees —una generalización directa del árbol AVL para múltiples hijos— precisamente porque garantizan búsquedas en $O(\log n)$ independientemente del tamaño de los datos. Una base de datos con mil millones de registros puede encontrar cualquier entrada en solo 30 comparaciones gracias a esta estructura. Sin árboles balanceados, Internet financiero —banca en línea, comercio electrónico, sistemas de pago— sería imposible a la escala que opera hoy.

**Los Grafos y sus algoritmos de recorrido** son el modelo matemático que subyace en Internet mismo. El algoritmo de enrutamiento que determina cómo cada paquete de datos viaja por Internet es un algoritmo de grafos. Los algoritmos de recomendación de plataformas como Netflix o Spotify modelan las preferencias de los usuarios como grafos y recorren sus vecinos para encontrar recomendaciones relevantes. Los sistemas de navegación GPS son algoritmos de camino más corto (Dijkstra) sobre grafos ponderados. En bioinformática, el ensamblaje de genomas es un problema de grafos. Los grafos no son una herramienta; son el lenguaje universal del modelado computacional.

### 6.2 Estrategias de Optimización Aplicadas

El análisis de los tres módulos revela patrones de optimización recurrentes que trascienden las estructuras individuales:

#### Optimización 1: Elección de representación interna

La decisión entre `list.pop(0)` y `collections.deque` para la Cola es paradigmática: la representación interna puede cambiar la complejidad de un mismo algoritmo de $O(n^2)$ a $O(n)$. Este principio se repite en grafos: la lista de adyacencia ($O(V + E)$) versus la matriz de adyacencia ($O(V^2)$) puede significar la diferencia entre procesar un grafo de un millón de nodos en segundos o en horas.

#### Optimización 2: Balanceo como inversión de rendimiento

El AVL invierte un pequeño costo constante por operación de inserción (las rotaciones, $O(1)$) para asegurar que todas las operaciones futuras sean $O(\log n)$. Esta es una decisión de diseño maestra: pagar un precio pequeño y constante ahora para evitar un costo potencialmente enorme en el futuro. El mismo principio está presente en los rehashes de tablas hash y los refactoring periódicos de código.

#### Optimización 3: Exploración determinista con ordenamiento de vecinos

La decisión de explorar vecinos en **orden alfabético** en BFS y DFS garantiza resultados reproducibles y deterministas, independientemente de la estructura interna del diccionario. Esto convierte un algoritmo potencialmente no determinista en uno perfectamente predecible para pruebas, auditorías y comparaciones.

#### Optimización 4: Separación de responsabilidades

En todos los sistemas implementados, la separación clara entre las estructuras de datos base (`Pila`, `Cola`, `Grafo`) y la lógica de negocio (`SistemaAtencionMedica`, `inventario BST`) permitió probar y verificar cada capa de forma independiente, reduciendo la superficie de errores y facilitando el mantenimiento.

### 6.3 Reflexión Crítica: Las Limitaciones que No Pueden Ignorarse

Una perspectiva académica honesta requiere reconocer las limitaciones:

- **Las estructuras implementadas en los laboratorios son didácticas, no de producción:** la Pila y la Cola de la semana 7 se implementaron sin librerías externas como restricción pedagógica, lo cual introduce la ineficiencia del `desencolar` en $O(n)$. En sistemas reales, `collections.deque` es la solución estándar.

- **El BST sin balanceo es inseguro para producción:** cualquier sistema que use un BST estándar sobre datos que pueden llegar en orden creciente (como IDs de base de datos generados automáticamente) está en riesgo de degradación severa. El AVL o el Red-Black Tree deben ser la elección predeterminada.

- **La recursión del DFS tiene límites prácticos:** Python tiene un límite de recursión predeterminado (~1000 llamadas). En grafos muy grandes o con cadenas largas, la versión recursiva del DFS puede lanzar un `RecursionError`. La solución es implementar DFS iterativo con una pila explícita.

- **Los grafos no ponderados tienen aplicaciones limitadas:** BFS encuentra el camino con menor número de aristas, pero en el mundo real (rutas con distancias, redes con costos) se requieren algoritmos como Dijkstra o Bellman-Ford sobre grafos ponderados, que no fueron cubiertos en estos laboratorios pero son el paso natural siguiente.

---

### 6.4 Conexiones Entre las Estructuras: Una Vista Unificada

Las cinco estructuras no son islas aisladas de conocimiento: están **profundamente interconectadas**:

1. **BFS usa una Cola:** el algoritmo de recorrido en anchura de un grafo (Semana 9) requiere exactamente la estructura estudiada en la Semana 7. Sin la Cola, BFS no puede garantizar el orden por niveles.

2. **DFS usa una Pila (implícita):** el DFS recursivo utiliza el *call stack* del sistema como pila implícita. Una implementación iterativa de DFS usa explícitamente la Pila de la Semana 7.

3. **Los árboles son grafos sin ciclos:** un Árbol BST o AVL es simplemente un grafo dirigido acíclico con una raíz designada. Los algoritmos de recorrido del árbol (preorden, inorden, postorden) son casos especiales de DFS.

4. **El historial del sistema médico usa Pila:** los resultados de la Semana 7 (Pila como historial de pacientes atendidos) aplican directamente como estructura de soporte para registrar el orden de visita en DFS.

Esta interconexión no es coincidencia: es reflejo de la unidad subyacente de la algorítmica. Las mismas ideas fundamentales —ordenar el acceso, garantizar completitud, optimizar el espacio— se manifiestan en diferentes formas a través de todas las estructuras.

## 7. Conclusión General

El estudio de las estructuras de datos y algoritmos de las semanas 7, 8 y 9 —Pilas, Colas, Árboles BST, Árboles AVL y Grafos con sus algoritmos BFS y DFS— representa mucho más que el aprendizaje de cinco herramientas técnicas independientes. Representa una **educación en el pensamiento algorítmico**: la capacidad de analizar un problema, identificar la estructura de datos que mejor modela sus características, implementarla de manera correcta y eficiente, y razonar sobre las consecuencias de esa elección a escala.

Las conclusiones específicas que emergen de este análisis integrado son las siguientes:

**Primera conclusión: La disciplina de acceso lo cambia todo.** La diferencia entre una Pila y una Cola no está en sus bits de implementación sino en la política de cuál elemento puede salir en cada momento. Esta simple decisión determina si un sistema puede modelar un historial de deshacer-rehacer (LIFO) o una cola de impresión (FIFO). El diseñador de software que domina estas políticas puede resolver problemas aparentemente complejos con herramientas elementales, como demostró el sistema de atención médica con prioridades implementado usando solo dos colas estándar.

**Segunda conclusión: El balance es la diferencia entre $O(\log n)$ y $O(n)$.** El árbol BST y el árbol AVL tienen exactamente la misma interfaz pública, el mismo uso desde el exterior, la misma propiedad de ordenamiento. Sin embargo, la presencia o ausencia del mecanismo de autobalanceo separa un sistema que garantiza $O(\log n)$ en todos los casos de uno que puede degradar a $O(n)$ ante datos ordenados. En producción, con millones de registros, esta diferencia es la que separa un sistema de bases de datos del mundo real de uno que se vuelve inutilizable. El AVL, y sus variantes como el Red-Black Tree, no son lujos académicos: son necesidades de ingeniería.

**Tercera conclusión: Los grafos son el modelo universal, y sus algoritmos de recorrido son las herramientas más versátiles del algoritmista.** BFS y DFS son, en esencia, la misma idea (recorrer todos los vértices y aristas alcanzables) implementada con diferentes estructuras auxiliares (cola vs. pila), produciendo diferentes órdenes de visita con diferentes propiedades de utilidad. BFS garantiza el camino más corto en grafos no ponderados; DFS es el fundamento de la detección de ciclos, los recorridos topológicos y la exploración por ramas. La complejidad $O(V + E)$ de ambos es óptima: no es posible explorar un grafo con menos trabajo que recorrer todos sus componentes.

**Cuarta conclusión: Las estructuras de datos se componen.** Los sistemas más poderosos no solo usan una estructura; las combinan. BFS usa una Cola. DFS usa una Pila (implícita en la recursión). El sistema de atención médica usa dos Colas más una Pila de historial. El diccionario Python que representa el grafo usa una lista (otra estructura lineal) para almacenar los vecinos de cada vértice. Esta composición es donde reside la verdadera ingeniería: no en usar cada estructura en aislamiento, sino en orquestarlas de manera que el conjunto resulte más poderoso que la suma de sus partes.

**Quinta conclusión: La complejidad algorítmica es el idioma del diseño escalable.** A lo largo de los tres módulos, el análisis de complejidad temporal y espacial fue la herramienta que permitió distinguir las soluciones eficientes de las ineficientes, predecir el comportamiento a escala, y justificar las decisiones de diseño. Un ingeniero que no domina la notación $O(\cdot)$ está diseñando a ciegas: puede producir código que funciona correctamente con datos pequeños y colapsa catastróficamente con datos reales.

En suma, las semanas 7, 8 y 9 de este curso no enseñaron cinco estructuras de datos; enseñaron **cinco maneras de pensar sobre la organización de la información**, cinco maneras de razonar sobre el tiempo y el espacio de computación, y cinco herramientas que, en manos de un ingeniero que las domina profundamente, son suficientes para diseñar los núcleos algorítmicos de la mayoría de los sistemas de software que el mundo real requiere.

La maestría en estructuras de datos no se adquiere leyendo sobre ellas: se forja implementándolas desde cero, analizando su comportamiento, descubriendo sus límites, y aprendiendo el lenguaje preciso con que describen la eficiencia. Este ensayo es un paso en ese camino, no su destino final.
